```python
# Import necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for plots
plt.style.use('ggplot')
sns.set_palette('husl')
```

```python
# Load the dataset

blinkit = pd.read_csv("blinkit_products.csv")

# Display first few rows

blinkit.head()
```

```python
# Data overview
print("\nMissing Values:")
print(blinkit.isnull().sum())
```

```python
# Descriptive statistics

print("Descriptive Statistics:")
blinkit.describe()
```

```python
# Product price distribution
plt.figure(figsize=(10,6))
sns.histplot(
    blinkit['price'],
    bins=20,
    kde=True,
    color='skyblue'
)
plt.title('Distribution of Product Prices')
plt.xlabel('Price (INR)')
plt.ylabel('Frequency')
plt.show()
```

***The graph shows that most products are concentrated in the low and medium price ranges, while only a few products are highly priced. This indicates that the platform mainly focuses on affordable grocery and daily-use products.***

```python
# Category performance
plt.figure(figsize=(12,6))
sns.countplot(x='category', data=blinkit)
plt.xticks(rotation=15)
plt.title('Product Category Distribution')
plt.xlabel('Category')
plt.ylabel('Number of Products')
plt.show()
```

***The category distribution graph shows that some categories contain a larger number of products compared to others. Categories with more products indicate higher product variety and potential customer demand.***

```python
# Inventory analysis
plt.figure(figsize=(10,6))
sns.scatterplot(
    x='min_stock_level',
    y='max_stock_level',
    data=blinkit
)
plt.title('Minimum vs Maximum Stock Levels')
plt.xlabel('Minimum Stock Level')
plt.ylabel('Maximum Stock Level')
plt.show()
```

***The scatter plot represents the relationship between minimum and maximum stock levels of products. Products with higher stock limits require better inventory management and storage planning.***

```python
# Shelf life analysis
plt.figure(figsize=(10,6))
sns.histplot(
    blinkit['shelf_life_days'],
    bins=15,
    kde=True
)
plt.title('Shelf Life Distribution')
plt.xlabel('Shelf Life (Days)')
plt.ylabel('Frequency')
plt.show()
```

***The shelf - life distribution helps understand product durability and storage requirements. Products with shorter shelf life require faster inventory movement and regular monitoring to reduce wastage.***

```python
# Profit margin analysis
plt.figure(figsize=(10,6))
sns.boxplot(y=blinkit['margin_percentage'])
plt.title('Margin Percentage Analysis')
plt.ylabel('Margin Percentage')
plt.show()
```

***The box plot shows the variation in profit margins across products. Some products generate higher profitability, while others provide lower margins, helping identify better-performing products.***

```python
# Price vs MRP comparison
plt.figure(figsize=(8,6))
sns.scatterplot(
    x='mrp',
    y='price',
    data=blinkit
)
plt.title('Price vs MRP')
plt.xlabel('MRP')
plt.ylabel('Selling Price')
plt.show()
```

***The scatter plot shows the relationship between Maximum Retail Price (MRP) and selling price. Most products are sold close to their MRP, while some products are available at discounted prices.***

```python
# Brand distribution
plt.figure(figsize=(12,6))
blinkit['brand'].value_counts().head(10).plot(kind='bar')
plt.title('Top Brands')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=25)
plt.show()
```

***The brand distribution graph highlights the brands with the highest number of products in the dataset. Brands with larger product counts may have stronger market presence and product variety.***
